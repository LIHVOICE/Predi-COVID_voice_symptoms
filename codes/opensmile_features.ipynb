```python
# libraries
import os
import pandas as pd
import numpy as np
from datetime import datetime
import opensmile
import librosa
```

```python
def opensmile_functionals(folder, sr=8000, feature_set='ComParE_2016'):
    # folder: store audios to be extracted
    # sr: sampling rate
    # feature_set: openSmile projects, should be ComParE_2016, GeMAPSv01b or eGeMAPSv02
    smile = opensmile.Smile(
        feature_set=getattr(opensmile.FeatureSet,feature_set),
        feature_level=opensmile.FeatureLevel.Functionals,
    )

    print(f'number of features: {len(smile.feature_names)}')
    features_opensmile = pd.DataFrame()
    for audio in os.listdir(folder):
        if audio.endswith('.wav'):
            print(audio)
            now = datetime.now()
            dt = now.strftime("%d/%m/%Y %H:%M:%S")
            print(dt)
            
            path_audio = os.path.join(folder, audio)
            y, sr = librosa.load(path=path_audio, sr=sr)
            f = smile.process_signal(signal=y, sampling_rate=sr)
            f_flat = pd.DataFrame(np.array(f))
            f_flat.columns = f.columns
            f_flat.insert(loc=0, column='file', value=audio)
            features_opensmile = features_opensmile.append(f_flat)
    return features_opensmile
```
